```python
def initialisation(m, n):
    # m nombre de lignes
    # n nombre de colonnes
    # retourne une matrice aléatoire (m, n+1)
    # avec une colonne biais (remplie de "1") tout à droite
    X = np.random.randn(m, n)
    X = np.concatenate((X, np.ones((X.shape[0], 1))), axis = 1)

    return X

initialisation(3, 4)
```

```python
from scipy import misc
import matplotlib
import matplotlib.pyplot as plt

face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
type(face)
face.shape
```

```python
face = face[::2, ::2]
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
h = face.shape[0]
w = face.shape[1]
print(h, w)
# on doit faire [h/4: -h/4]
# double division entière
zoom_face = face[h//4: -h//4, w//4: -w//4]
zoom_face[zoom_face > 150] = 255
plt.imshow(zoom_face, cmap=plt.cm.gray)
plt.show()
```

Z-score standardisation

```python
np.random.seed(0)
A = np.random.randint(0, 100, [10, 5])
A
```

```python
D = (A - A.mean(axis=0)) / A.std(axis=0)
D
```

```python
D.mean(axis=0)
```

```python
D.std(axis=0)
```
